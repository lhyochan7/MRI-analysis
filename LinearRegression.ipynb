```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as pltt
from scipy import stats
```

```python
def get_df_name(df):
    name =[x for x in globals() if globals()[x] is df][0]
    return name
```

```python
df = pd.read_csv('aqua_test.csv')

CN = df.loc[(df['Group'] == 'CN')] 


MCI = df.loc[(df['Group'] == 'MC')]


AD = df.loc[(df['Group'] == 'AD')]

groups = [CN, MCI, AD]
features = ['left_hippocampus', 'right_hippocampus', 'left_cerebral_white_matter', 'right_cerebral_white_matter']

#print(features)
#print(CN_M)
```

```python
for group in groups:
    for feature in features:
        #print(feature)
        X = group['age'].values.reshape(-1, 1)  # values converts it into a numpy array
        Y_r = group[feature].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column

        # create object for class
        lin_reg_r = LinearRegression() 

        # perform linear regression
        lin_reg_r.fit(X, Y_r)  

        Y_pred_r = lin_reg_r.predict(X)

        plt.scatter(X, Y_r)

        plt.plot(X, Y_pred_r, color='red')

        plt.title('%s = %s' %(get_df_name(group), feature), loc='right')
        plt.xlabel('Age')
        plt.ylabel('Volume')
    
        plt.show()

```
