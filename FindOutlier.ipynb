```python
data = pd.read_csv('aqua_volume_1yr.csv')

#print(df)

#exception = ['Left-WM-hypointensities', 'Right-WM-hypointensities', 'Left-non-WM-hypointensities', 'Right-non-WM-hypointensities']
column_ls = []


for column in data.columns:
    column_ls.append(column)

#print(column_ls)
df = data.loc[:,column_ls]
    
    
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df_sum = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()

#print(df_sum)

temp = []

for i in df_sum:
    temp.append(i)
    #print(i)

count = 0
for j in df_sum.index:
    print("%s = %d" %(j,temp[count]))
    count+=1
    
# Plot Boxplot
#print(df.columns)

for c in df.columns:
    if(c == 'ID'):
        continue
    plt.figure(figsize=(2, 5))
    boxplot = df.boxplot(column = [c])
    plt.show()



#print(column_ls)
#print(df)
```
